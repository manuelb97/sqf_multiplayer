```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = [80, 70, 60, 50, 40, 30]
y = [5,5,5,15,35,60]

z = np.polyfit(x, y, 2)
p = np.poly1d(z)

xp = np.linspace(min(x), max(x), 1000)
plt.plot(x, y, '.', xp, p(xp), '-')

for idx, x in enumerate(xp):
    if idx % 100 == 0 or idx == 999:
        print(x, p(x))
```
